**Array Basics**

1.Attributes of an array
Introduce attributes that describe the array .
Attributes: .shape,.ndim,.size,.dtype

```python
import numpy as np
arr = np.array([[1,2,3,],[4,5,6]])

print("which dimensional of array :- ",arr.shape)
print(arr.ndim)
print(arr.size)
print(arr.dtype)
```

Indexing and Slicing
Now we have indexing and slicing aswell and for that we are going to see 1d indexing lets start with 1 d indexing and it is similar to list.

```python
arr = np.array([1,2,3,4,5])
print(arr[0])
print(arr[1])
```

Now lets see 2D indexing

```python
arr = np.array(([2,3,5],[7,8,9],[1,5,8]))
arr
print(arr[0])
```

Now we can also do the slicing lets see the slicing in 1D

```python
a = np.array([1,2,3,4,5,6,7])
a[:3]
```

lets see the slicing in 2D

```python
arr = np.array([[1,2,3],[6,4,5],[6,7,8],[3,9,10]])
print(arr)
arr[0:2,0:2]
```

There is one more thing as boolean indexing

```python
arr = np.array([5,10,20,30,40])
print(arr > 12)
print(arr[arr > 12])
```
