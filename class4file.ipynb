##Fancy Indexing
you can Pick elements from specific positions using a list/array of indices.

```python
import numpy as np
arr = np.array([1,20,30,40,50])
arr[[0,2,4]]
```

You can do this fancy indexing in 2D Matrix aswell

```python
matrix = np.arange(1,13).reshape(3,4)
print(matrix)
matrix[[0,1,2],[1,2,3]]
```

##Conditional Selection
Find values based on conditions or replace them.

```python
arr = np.array([4,12,34,87,6])
arr[arr > 10]
np.where(arr>10)
np.where(arr>10,arr,0)
```

##Sorting & Ordering
np.sort() → returns sorted array

np.argsort() → returns indices of sorted order

```python
arr = np.array([4,12,34,87,6])
np.sort(arr)
```

```python
np.argsort(arr)
```

##Concatenation & Stacking

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
np.concatenate([a,b])
```

Vertical & Horizontal Stacking

```python
A = np.array([[1,2],[4,9]])
B = np.array([[5,6],[7,8]])
```

```python
np.vstack([A,B])
```

```python
np.hstack([B,A])
```
